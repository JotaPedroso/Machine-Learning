```python
from sklearn import datasets
from sklearn.neural_network import MLPClassifier ## Classificador do Multi-Layer Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score ## metrica de acerto
import matplotlib.pyplot as plt # biblioteca para geração de grafico
```

```python
iris = datasets.load_iris()
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(iris.data, iris.target,
                                                                  test_size=0.3, random_state=0)
```

```python
modelo = MLPClassifier(
    verbose=True,
    hidden_layer_sizes=(5,4),      #Camadas ocultas e qtt de neuronios em cada
    activation='relu',             #função de ativação padrão
    batch_size=20,
    learning_rate='adaptive',
    momentum=0.9,
    early_stopping=False,
    max_iter=1000,                  #qtt de epochs
    random_state=10,

)
modelo.fit(X_treinamento, y_treinamento)
```

```python
plt.plot(modelo.loss_curve_) # Biblioteca para criação do gráfico
plt.xlabel("Iteraçoes")
plt.ylabel("Valor de loss")
plt.title("Loss")
plt.show()
```

```python
previsoes = modelo.predict(X_teste)
previsoes
```

```python
print(accuracy_score(y_teste,previsoes))
```
